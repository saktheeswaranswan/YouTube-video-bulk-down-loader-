```python
import os
import imageio_ffmpeg as ffmpeg
from moviepy.editor import VideoFileClip, clips_array

# Define the path to the folder containing your videos
video_folder = '/content/videos/'

# Get a list of all .mp4 files in the folder
video_files = [f for f in os.listdir(video_folder) if f.endswith('.mp4')]

# Create lists to store video clips for 8K and 4K resolutions
video_clips_8k = []
video_clips_4k = []

# Load each video file and append it to the appropriate list based on resolution
for video_file in video_files:
    video_path = os.path.join(video_folder, video_file)
    video_clip = VideoFileClip(video_path)

    # Resize the video to 8K (if it's 8K resolution) or 4K (if it's 4K resolution)
    if video_clip.size == (7680, 4320):  # Assuming 8K resolution is 7680x4320
        video_clips_8k.append(video_clip)
    elif video_clip.size == (3840, 2160):  # Assuming 4K resolution is 3840x2160
        video_clips_4k.append(video_clip)

# Concatenate the video clips horizontally for each resolution
final_video_8k = clips_array([video_clips_8k])
final_video_4k = clips_array([video_clips_4k])

# Define the output file paths for the merged videos
output_path_8k = '/content/merged_video_8k.mp4'
output_path_4k = '/content/merged_video_4k.mp4'

# Write the merged videos to the output files using imageio_ffmpeg with GPU acceleration
final_video_8k.write_videofile(output_path_8k, codec='libx264', threads=4, preset='fast', ffmpeg_params=['-hwaccel', 'auto'])
final_video_4k.write_videofile(output_path_4k, codec='libx264', threads=4, preset='fast', ffmpeg_params=['-hwaccel', 'auto'])

print(f"Merged 8K video saved to {output_path_8k}")
print(f"Merged 4K video saved to {output_path_4k}")
```
